```python
import pandas as pd
import numpy as np
```

```python
df_simulation_store_product = pd.read_json('simulation_store_product.json', lines = True)
```

```python
df_simulation_store_product_agg = df_simulation_store_product[
    (df_simulation_store_product['LUNAR_DAY'] >= 21)
    & (df_simulation_store_product['VALUE_ACTUAL_BASE_SALE_QTY'] > 0)
].groupby(['CITY', 'STORE_ID', 'PRODUCT_ID'])[['VALUE_ACTUAL_BASE_SALE_QTY', 'VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityLunarDay', 'VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions']].sum().reset_index().sort_values(by = ['VALUE_ACTUAL_BASE_SALE_QTY'], ascending = False)

df_simulation_store_product_agg['MAPE_byMCH5CityLunarDay'] = abs(df_simulation_store_product_agg['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityLunarDay'] - df_simulation_store_product_agg['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_store_product_agg['VALUE_ACTUAL_BASE_SALE_QTY'] 
df_simulation_store_product_agg['MAPE_byMCH5CityStoreTransactions'] = abs(df_simulation_store_product_agg['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions'] - df_simulation_store_product_agg['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_store_product_agg['VALUE_ACTUAL_BASE_SALE_QTY'] 

df_simulation_store_product_agg
```

```python
df_simulation_store_product_agg_top100stores = df_simulation_store_product_agg[df_simulation_store_product_agg['STORE_ID'].isin(df_simulation_store_product_agg.groupby(['STORE_ID'])['VALUE_ACTUAL_BASE_SALE_QTY'].sum().reset_index().sort_values(by = ['VALUE_ACTUAL_BASE_SALE_QTY'], ascending = False)[:100]['STORE_ID'].unique())]
df_simulation_store_product_agg_top100stores

```

```python
df_simulation_store_product_agg_top100stores['MAPE_byMCH5CityLunarDay'] = abs(df_simulation_store_product_agg_top100stores['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityLunarDay'] - df_simulation_store_product_agg_top100stores['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_store_product_agg_top100stores['VALUE_ACTUAL_BASE_SALE_QTY']

df_simulation_store_product_agg_top100stores['MAPE_byMCH5CityStoreTransactions'] = abs(df_simulation_store_product_agg_top100stores['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions'] - df_simulation_store_product_agg_top100stores['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_store_product_agg_top100stores['VALUE_ACTUAL_BASE_SALE_QTY']


```

```python
df_forecast_agg_city_product_daily = pd.read_json('simulation_agg_city_product.json', lines = True)
df_forecast_agg_city_product_daily
```

```python
df_simulation_city_product_agg = df_forecast_agg_city_product_daily[
    (df_forecast_agg_city_product_daily['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions'] > 0)
    & (df_forecast_agg_city_product_daily['VALUE_ACTUAL_BASE_SALE_QTY'] > 0)
].groupby(['CITY', 'PRODUCT_ID'])[[
    'VALUE_ACTUAL_BASE_SALE_QTY', 'VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityLunarDay', 'VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions'
]].sum().reset_index().sort_values(by= ['VALUE_ACTUAL_BASE_SALE_QTY'], ascending = False)

df_simulation_city_product_agg['MAPE_byMCH5CityLunarDay'] = abs(df_simulation_city_product_agg['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityLunarDay'] - df_simulation_city_product_agg['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_city_product_agg['VALUE_ACTUAL_BASE_SALE_QTY'] 
df_simulation_city_product_agg['MAPE_byMCH5CityStoreTransactions'] = abs(df_simulation_city_product_agg['VALUE_FORECAST_BASE_SALE_QTY_byMCH5CityStoreTransactions'] - df_simulation_city_product_agg['VALUE_ACTUAL_BASE_SALE_QTY']) / df_simulation_city_product_agg['VALUE_ACTUAL_BASE_SALE_QTY'] 

df_simulation_city_product_agg
```
